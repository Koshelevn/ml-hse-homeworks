**Доп (0.25 балла):**
* <font color='#9933CC'>Вместо удаления признак `torque` разделите на два: собственно `torque` и `max_torque_rpm`.</font> Учтите единицы измерения. Они разные ☹

```python
import re

df_train.loc[:, ['mileage', 'engine', 'max_power']] = df_train.loc[:, ['mileage', 'engine', 'max_power']].applymap(lambda x: re.sub(r"[^0-9.]", "", str(x)) or None).astype('float')
df_test.loc[:, ['mileage', 'engine', 'max_power']] = df_test.loc[:, ['mileage', 'engine', 'max_power']].applymap(lambda x: re.sub(r"[^0-9.]", "", str(x)) or None).astype('float')
```

```python
def matcher(torque):
  if isinstance(torque, float):
    return [torque]
  torque = torque.lower()
  numbers = re.findall(r"([0-9,.]+)+", torque)
  numbers = list(map(lambda x: float(x.replace(',', '')), numbers))
  if 'kgm' in torque:
    numbers[0] = numbers[0]*9.8 # Переводим в nm
  if (len(numbers) == 1):
    if ('kgm' in torque) or ('nm' in torque): # Если указан только torque без rpm
      return numbers[0], np.nan
    else:
      return np.nan, numbers[1]
  return numbers[0], numbers[-1]


df_test['max_torque_rpm'] = df_test.torque.apply(lambda x: matcher(x)[-1])
df_test['torque'] = df_test.torque.apply(lambda x: matcher(x)[0])

df_train['max_torque_rpm'] = df_train.torque.apply(lambda x: matcher(x)[-1])
df_train['torque'] = df_train.torque.apply(lambda x: matcher(x)[0])
```

```python
from datetime import datetime

def df_processor(df: pd.DataFrame):
    df['driven_per_year'] = df.km_driven/(datetime.now().year - df.year)
    df['power_per_liter'] = df.max_power / df.engine
    df['year_squared'] = df.year**2
    df['rpm_per_torque'] = df.max_torque_rpm / df.torque
    df['brand'] = df.name.apply(lambda x: x.split(' ')[0])
    df.drop(['name'], axis=1, inplace=True)

    return df

X_train_fe = df_processor(X_train_fe)
X_test_fe = df_processor(X_test_fe)
```

```python
from sklearn.preprocessing import OneHotEncoder # или можно использовать get_dummies из библиотеки pandas

cat = ['brand', 'fuel', 'seller_type', 'transmission', 'owner', 'seats']

encoder = OneHotEncoder(drop='first', handle_unknown='ignore')

transformed_train = encoder.fit_transform(X_train_fe[cat])
transformed_train = pd.DataFrame.sparse.from_spmatrix(transformed_train, columns=encoder.get_feature_names_out())
X_train_fe = pd.concat([transformed_train, X_train_fe], axis=1).drop(cat, axis=1)

transformed_test = encoder.transform(X_test_fe[cat])
transformed_test = pd.DataFrame.sparse.from_spmatrix(transformed_test, columns=encoder.get_feature_names_out())
X_test_fe = pd.concat([transformed_test, X_test_fe], axis=1).drop(cat, axis=1)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_fe)
X_test_scaled = scaler.transform(X_test_fe)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

params = {'alpha': (np.logspace(-10, 10, 10)), 'tol': (np.linspace(0.00001, 1, 10))}
ridge = Ridge()
ridge_search = GridSearchCV(ridge, params, cv = 10, scoring='r2')
ridge_search.fit(X_train_scaled, np.log(y_train))

ridge_grid_model = Ridge(**ridge_search.best_params_)
ridge_grid_model.fit(X_train_scaled, np.log(y_train))

train_preds = ridge_grid_model.predict(X_train_scaled)
test_preds = ridge_grid_model.predict(X_test_scaled)

print(f'Train:\nMSE = {MSE(y_train, np.exp(train_preds))}, r2 = {r2_score(y_train, np.exp(train_preds))}')
print(f'Test:\nMSE = {MSE(y_test, np.exp(test_preds))}, r2 = {r2_score(y_test, np.exp(test_preds))}')
```
